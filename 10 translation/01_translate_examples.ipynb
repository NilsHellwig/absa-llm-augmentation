```python
from deep_translator import GoogleTranslator
import os, sys
import json
```

```python
sys.path.append(os.path.dirname("../04 llm synthesis/"))
from helper_synthesis import convert_ner_to_xml, xml_to_json
```

```python
#GoogleTranslator(source='auto', target='de').translate(sentence)
```

```python
for i in range(1, 6):
    backtranslation_split = []
    with open(f'../07 train models/real/split_{i}.json', 'r') as json_datei:
        real_split = json.load(json_datei)
        for idx, example in enumerate(real_split):
            print(idx, example["id"])
            example_with_tags_german = convert_ner_to_xml(example)
            example_with_tags_english = GoogleTranslator(
                source='de', target='en').translate(example_with_tags_german)
            back_translation_german = GoogleTranslator(
                source='en', target='de').translate(example_with_tags_english)
            back_translation_german = back_translation_german.replace(
                "polarity =", "polarity=")
            back_translation_json = "not-in-label"
            count = 0
            while back_translation_json == "not-in-label":
                if count < 20:
                    try:
                        back_translation_json = xml_to_json(back_translation_german, [(
                            aspect["label"], aspect["polarity"]) for aspect in example["tags"]], "de", i, check_label=True)
                    except:
                        pass
                else:
                    back_translation_json = xml_to_json(example_with_tags_german, [(
                        aspect["label"], aspect["polarity"]) for aspect in example["tags"]], "de", i, check_label=True)
                    break
                count+=1

            bt_example = example.copy()
            bt_example["text"] = back_translation_json["text"]
            bt_example["tags"] = back_translation_json["tags"]

            backtranslation_split.append(bt_example)

    with open(f'../07 train models/real_bt/split_{i}.json', "w", encoding="utf-8") as outfile:
        json.dump(backtranslation_split, outfile, ensure_ascii=False)
```
